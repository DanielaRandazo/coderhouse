# Hip√≥tesis Univariada 1:
La calificaci√≥n promedio de las pel√≠culas en el conjunto de datos es superior a 6 (averageRating).


#Conclusion:

Tras analizar la distribuci√≥n de las calificaciones promedio de las pel√≠culas en nuestro conjunto de datos, hemos encontrado que la mayor√≠a de las pel√≠culas tienen calificaciones superiores a 6. Este resultado respalda nuestra hip√≥tesis inicial.

```python
import matplotlib.pyplot as plt


calificaciones_peliculas = DF_IMDB['averageRating']

# Creo un histograma
plt.hist(calificaciones_peliculas, bins=20, color='blue', alpha=0.7)
plt.axvline(x=6, color='red', linestyle='dashed', linewidth=2, label='Calificaci√≥n Promedio >= 6')

# Etiquetas y t√≠tulo
plt.xlabel('Calificaci√≥n Promedio')
plt.ylabel('N√∫mero de Pel√≠culas')
plt.title('Distribuci√≥n de Calificaciones de Pel√≠culas')
plt.legend()

# Muestrohiijajaj el gr√°fico
plt.show()

```

# Hipotesis Bivariada 1:
Las pel√≠culas de g√©nero "Acci√≥n" tienen una calificaci√≥n promedio significativamente diferente de las pel√≠culas de g√©nero "Drama".

#Conclusion:
Despu√©s de analizar las calificaciones promedio de las pel√≠culas de g√©nero 'Acci√≥n' y 'Drama', no encontramos evidencia que respalde la hip√≥tesis de que estas dos categor√≠as de pel√≠culas tienen calificaciones promedio significativamente diferentes. Ambos g√©neros muestran calificaciones promedio similares, lo que sugiere que los espectadores y/o cr√≠ticos eval√∫an estas pel√≠culas de manera comparable en t√©rminos de calidad. Por lo tanto, no podemos concluir que haya una diferencia significativa en las calificaciones promedio entre las pel√≠culas de g√©nero 'Acci√≥n' y 'Drama' en nuestro conjunto de datos.


```python
import matplotlib.pyplot as plt
import pandas as pd


# Filtro las pel√≠culas de g√©nero "Acci√≥n" y "Drama"
accion_movies = DF_IMDB[DF_IMDB['genres'].str.contains('Action')]
drama_movies = DF_IMDB[DF_IMDB['genres'].str.contains('Drama')]

# Creo un gr√°fico de cajas comparando las calificaciones promedio de ambos grupos
plt.boxplot([accion_movies['averageRating'], drama_movies['averageRating']], labels=['Acci√≥n', 'Drama'])

# Etiquetas y t√≠tulo
plt.xlabel('G√©nero')
plt.ylabel('Calificaci√≥n Promedio')
plt.title('Comparaci√≥n de Calificaciones Promedio entre Pel√≠culas de Acci√≥n y Drama')

# Muestro el gr√°fico
plt.show()
```

# Hipotesis Bivariada 2:

La distribuci√≥n de g√©neros cinematogr√°ficos ha evolucionado a lo largo de los a√±os, y ciertos g√©neros son m√°s comunes en ciertas d√©cadas.

#Conclusion:
Despu√©s de un amplio an√°lisis de la evoluci√≥n de los g√©neros cinematogr√°ficos a lo largo de las d√©cadas, hemos llegado a la conclusi√≥n de que la distribuci√≥n de g√©neros se ha mantenido relativamente constante en el tiempo. Si bien ha habido fluctuaciones y cambios menores en la popularidad de g√©neros espec√≠ficos en diferentes d√©cadas, en general, no hemos observado una tendencia clara hacia la dominaci√≥n de un g√©nero particular en una d√©cada en particular.

Nuestros gr√°ficos y datos revelan que la producci√≥n cinematogr√°fica ha experimentado un crecimiento significativo a lo largo de las d√©cadas, abarcando una amplia variedad de g√©neros. Esto sugiere que la industria cinematogr√°fica ha sido diversa y ha crecido significativamente a lo largo del tiempo.

En resumen, la evoluci√≥n de los g√©neros ha ido en crecimiento constante y parejo a lo largo de las decadas.



```python
import matplotlib.pyplot as plt
import pandas as pd

# Agrupo el conjunto de datos por d√©cada y tipos y cuento la cantidad de pel√≠culas en cada grupo.
DF_IMDB['decade'] = (DF_IMDB['startYear'] // 10) * 10  # Agrupo por d√©cadas
genre_decade_counts = DF_IMDB.groupby(['decade', 'titleType']).size().unstack(fill_value=0)

# Gr√°fico de barras apiladas para la evoluci√≥n de tipos de pelis a lo largo de las d√©cadas.
plt.figure(figsize=(20, 8))
genre_decade_counts.plot(kind='bar', stacked=True)
plt.xlabel('D√©cada de Inicio')
plt.ylabel('Cantidad de Pel√≠culas')
plt.title('Evoluci√≥n de Tipos de peliculas por D√©cada de Inicio')
plt.legend(title='Tipos de Pelicula', bbox_to_anchor=(1, 1))

# Muestro el gr√°fico
plt.show()
```

El idioma en el que se realiza una pel√≠cula, su calificaci√≥n promedio y su tipo de t√≠tulo (por ejemplo, "Pel√≠cula" o "Serie de TV") est√°n relacionados y muestran patrones distintos.

#Conclusion:

Despues de analizar el conjunto de datos y representarlo gr√°ficamente, observamos que todas las pel√≠culas registradas est√°n en idioma ingl√©s. No hemos encontrado pel√≠culas en otros idiomas en nuestro conjunto de datos. Esto indica una fuerte concentraci√≥n de pel√≠culas en ingl√©s y la ausencia de pel√≠culas en otros idiomas en nuestra muestra.
```python
import seaborn as sns
import matplotlib.pyplot as plt


# Creoun gr√°fico para representar la relaci√≥n entre idioma, calificaci√≥n promedio y tipo de t√≠tulo.
plt.figure(figsize=(12, 8))
sns.scatterplot(data=DF_IMDB, x='averageRating', y='language', hue='titleType', palette='viridis', s=100)
plt.xlabel('Calificaci√≥n Promedio')
plt.ylabel('Idioma')
plt.title('Relaci√≥n entre Idioma, Calificaci√≥n Promedio y Tipo de T√≠tulo')
plt.legend(title='Tipo de T√≠tulo', loc='upper right')

# Muestro el gr√°fico
plt.show()

```
