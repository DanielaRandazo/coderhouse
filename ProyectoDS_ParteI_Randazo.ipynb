# Hipótesis Univariada 1:
La calificación promedio de las películas en el conjunto de datos es superior a 6 (averageRating).


#Conclusion:

Tras analizar la distribución de las calificaciones promedio de las películas en nuestro conjunto de datos, hemos encontrado que la mayoría de las películas tienen calificaciones superiores a 6. Este resultado respalda nuestra hipótesis inicial.
La evidencia del gráfico sugiere que las películas en general tienden a recibir calificaciones favorables por parte de los usuarios o críticos, lo que indica un nivel de satisfacción por encima de la media.

```python
import matplotlib.pyplot as plt


calificaciones_peliculas = DF_IMDB['averageRating']

# Creo un histograma
plt.hist(calificaciones_peliculas, bins=20, color='blue', alpha=0.7)
plt.axvline(x=6, color='red', linestyle='dashed', linewidth=2, label='Calificación Promedio >= 6')

# Etiquetas y título
plt.xlabel('Calificación Promedio')
plt.ylabel('Número de Películas')
plt.title('Distribución de Calificaciones de Películas')
plt.legend()

# Muestrohiijajaj el gráfico
plt.show()

```

Las películas producidas en la región de Estados Unidos claramente lideran en términos de número de votos a nivel mundial.
```python
# Datos
region = DF_IMDB['region']
votos = DF_IMDB['numVotes']

# Creo el objeto
fig, ax = plt.subplots()

# Crea el gráfico de barras
ax.bar(region, votos)

# Agregar etiquetas y título
ax.set_xlabel('Region')
ax.set_ylabel('Votos')
ax.set_title('Votos de Peliculas segun Region')

# Muestro el gráfico
plt.show()
```

# Hipotesis Bivariada 1:
Las películas de género "Acción" tienen una calificación promedio significativamente diferente de las películas de género "Drama".

#Conclusion:
Después de analizar las calificaciones promedio de las películas de género 'Acción' y 'Drama', no encontramos evidencia que respalde la hipótesis de que estas dos categorías de películas tienen calificaciones promedio significativamente diferentes. Ambos géneros muestran calificaciones promedio similares, lo que sugiere que los espectadores y/o críticos evalúan estas películas de manera comparable en términos de calidad. Por lo tanto, no podemos concluir que haya una diferencia significativa en las calificaciones promedio entre las películas de género 'Acción' y 'Drama' en nuestro conjunto de datos.


```python
import matplotlib.pyplot as plt
import pandas as pd


# Filtro las películas de género "Acción" y "Drama"
accion_movies = DF_IMDB[DF_IMDB['genres'].str.contains('Action')]
drama_movies = DF_IMDB[DF_IMDB['genres'].str.contains('Drama')]

# Creo un gráfico de cajas comparando las calificaciones promedio de ambos grupos
plt.boxplot([accion_movies['averageRating'], drama_movies['averageRating']], labels=['Acción', 'Drama'])

# Etiquetas y título
plt.xlabel('Género')
plt.ylabel('Calificación Promedio')
plt.title('Comparación de Calificaciones Promedio entre Películas de Acción y Drama')

# Muestro el gráfico
plt.show()
```

# Hipotesis Bivariada 2:

La distribución de géneros cinematográficos ha evolucionado a lo largo de los años, y ciertos géneros son más comunes en ciertas décadas.

#Conclusion:
Después de un amplio análisis de la evolución de los géneros cinematográficos a lo largo de las décadas, hemos llegado a la conclusión de que la distribución de géneros se ha mantenido relativamente constante en el tiempo. Si bien ha habido fluctuaciones y cambios menores en la popularidad de géneros específicos en diferentes décadas, en general, no hemos observado una tendencia clara hacia la dominación de un género particular en una década en particular.

Nuestros gráficos y datos revelan que la producción cinematográfica ha experimentado un crecimiento significativo a lo largo de las décadas, abarcando una amplia variedad de géneros. Esto sugiere que la industria cinematográfica ha sido diversa y ha crecido significativamente a lo largo del tiempo.

En resumen, la evolución de los géneros ha ido en crecimiento constante y parejo a lo largo de las decadas.



```python
import matplotlib.pyplot as plt
import pandas as pd

# Agrupo el conjunto de datos por década y tipos y cuento la cantidad de películas en cada grupo.
DF_IMDB['decade'] = (DF_IMDB['startYear'] // 10) * 10  # Agrupo por décadas
genre_decade_counts = DF_IMDB.groupby(['decade', 'titleType']).size().unstack(fill_value=0)

# Gráfico de barras apiladas para la evolución de tipos de pelis a lo largo de las décadas.
plt.figure(figsize=(20, 8))
genre_decade_counts.plot(kind='bar', stacked=True)
plt.xlabel('Década de Inicio')
plt.ylabel('Cantidad de Películas')
plt.title('Evolución de Tipos de peliculas por Década de Inicio')
plt.legend(title='Tipos de Pelicula', bbox_to_anchor=(1, 1))

# Muestro el gráfico
plt.show()
```

# Hipotesis Trivariada 1:

El idioma en el que se realiza una película, su calificación promedio y su tipo de título (por ejemplo, "Película" o "Serie de TV") están relacionados y muestran patrones distintos.

#Conclusion:

Despues de analizar el conjunto de datos y representarlo gráficamente, observamos que todas las películas registradas están en idioma inglés. No hemos encontrado películas en otros idiomas en nuestro conjunto de datos. Esto indica una fuerte concentración de películas en inglés y la ausencia de películas en otros idiomas en nuestra muestra. Por lo tanto, no podemos analizar las relaciones entre el idioma, la calificación promedio y el tipo de título en idiomas distintos al inglés debido a la falta de datos en esos idiomas en nuestro conjunto.

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Creoun gráfico para representar la relación entre idioma, calificación promedio y tipo de título.
plt.figure(figsize=(12, 8))
sns.scatterplot(data=DF_IMDB, x='averageRating', y='language', hue='titleType', palette='viridis', s=100)
plt.xlabel('Calificación Promedio')
plt.ylabel('Idioma')
plt.title('Relación entre Idioma, Calificación Promedio y Tipo de Título')
plt.legend(title='Tipo de Título', loc='upper right')

# Muestro el gráfico
plt.show()

```
